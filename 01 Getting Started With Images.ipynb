# GETTING STARTED WITH IMAGES

````python
import cv2
import numpy as np
````

# Read an image

Syntax ```cv2.imread(image_path, flag)```

**flags**
*  1 - Loads a color image
*  0 - Loads image in grayscale mode
* -1 - Loads image as such including alpha channel

````python
# Load an image in grayscale
img = cv2.imread('dog.jpg', 0)
````

# Display an image

````python
# This method is used in cases when window is already created n image is loaded afterwards.
cv2.namedWindow('image', cv2.WINDOW_NORMAL)

# Display an image
cv2.imshow('image',img)

''' 
WaitKey ---
If argument is 0, wait indefinitely until any key is pressed.
If non-zero, waits for specified miliseconds only. After that it destroys automatically. 
Or you can destroy before the specfied time by pressing any key.
''' 
cv2.waitKey(0)                 
cv2.destroyAllWindows()
````

# Write an image

````python
cv2.imwrite('dog_gray.jpg', img)
````

# Using Matplotlib

````python
import matplotlib.pyplot as plt

img = cv2.imread('dog.jpg', 0)

plt.imshow(img, cmap='gray', interpolation='bicubic')
# plt.xticks([]), plt.yticks([])      # to hide tick values on X & Y axis.
plt.show()
````
